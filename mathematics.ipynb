```python
from sympy import *

x, y = symbols('x y')

# derivative for first function
# need to underscore y to prevent variable clash
_y = x**2 + 1
dy_dx = diff(_y)

# derivative for second function
z = y**3 - 2
dz_dy = diff(z)

# Calculate derivative with and without
# chain rule, substitute y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

# Prove chain rule by showing both are equal
print(dz_dx_chain) 
print(dz_dx_no_chain) 
```

```python
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0

    for i in range(1, n + 1):
       midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
       total_sum += f(midpoint)
    
    return total_sum * delta_x

def my_function(x):
    return x**2 + 1

area = approximate_integral(a=0, b=1, n=5, f=my_function)

print(area)
```

```python
area = approximate_integral(a=0, b=1, n=1000, f=my_function)

print(area)
```

```python
area = approximate_integral(a=0, b=1, n=1_000_000, f=my_function)

print(area)
```

```python
from sympy import *

# Declare variables to SymPy
x, i, n = symbols('x i n')

# Declare function and range
f = x**2 + 1
lower, upper = 0, 1

# Calculate width and each rectangle height at index "i"
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)

# Iterate all "n" rectangles and sum their areas
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

# Calculate the area by approaching the number of rectangles "n" to infinity
area = limit(n_rectangles, n, oo)

print(area) # prints 4/3

```

```python
# bayes theorem

# P(A|B) = P(B|A) * P(A) / P(B)
# apply this to those that have cancer who drink coffee

p_coffee_drinker = .65
p_cancer = .005
p_coffee_drinker_given_cancer = .85

p_cancer_given_coffee_drinker = p_coffee_drinker_given_cancer * p_cancer / p_coffee_drinker

print(p_cancer_given_coffee_drinker)
```

## Binomial Distribution

```python
from scipy.stats import binom
 
#Here, we see the probability of k successes for each bar out of 10 trials. This binomial distribution assumes a probability p of 90%, meaning there is a .90 (or 90%) chance for a success to occur

n = 10
p = .90

for k in range(n + 1):
    probability = binom.pmf(k, n, p)
    print("{} - {}".format(k, probability))
```

```python
# a chart of the beta distribution 
# calculate area under the curve up to 90%
from scipy.stats import beta

a = 8
b = 2

p = beta.cdf(.90, a, b)

print(p)
```

```python
# plot the beta distribution
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = beta.pdf(x, a, b)
# shade in the auc up to 90%
plt.fill_between(x, y, where=x < .90)
plt.plot(x, y)
plt.show()
```

```python
# If we want to find the opposite probability of an event (greater than 0.90 as 
# opposed to less than 0.90), just subtract the probability of being less than 0.90 
# from 1.0, and the remaining probability will capture being greater than 0.90.

from scipy.stats import beta

a = 8
b = 2

p = 1 - beta.cdf(.90, a, b)

print(p)
```

```python
#There is a 30% chance of rain today, and a 40% chance your umbrella order 
# will arrive on time. You are eager to walk in the rain today and cannot do # so without either! What is the probability it will rain AND your umbrella 
# will arrive?

p_rain = .30
p_umbrella = .40

p_rain_and_umbrella = p_rain * p_umbrella

print(p_rain_and_umbrella)
```

```python
# There is a 30% chance of rain today, and a 40% chance your umbrella order
#  will arrive on time.  You will be able to run errands only if it does not
#  rain or your umbrella arrives. What is the probability it will not rain OR 
# your umbrella arrives?

p_rain = .30
p_umbrella = .40

# create union probability
p_rain_or_umbrella = (1 - p_rain) + p_umbrella -((1 - p_rain) * p_umbrella)

print(p_rain_or_umbrella)
```

```python
# There is a 30% chance of rain today, and a 40% chance your umbrella order 
# will arrive on time.
# However, you found out if it rains there is only a 20% chance your umbrella 
# will arrive on time.
# What is the probability it will rain AND your umbrella will arrive on time?

p_rain = .30
p_umbrella_given_rain = .20

p_rain_and_umbrella = p_rain * p_umbrella_given_rain

print(p_rain_and_umbrella)
```

```python
# You have 137 passengers booked on a flight from Las Vegas to Dallas. 
# However, it is Las Vegas on a Sunday morning and you estimate each 
# passenger is 40% likely to not show up.

from scipy.stats import binom
n = 137
p = .40

# calculate the probability of no shows
p_50_or_more_no_shows = 0

for x in range (50, 138):
    p_50_or_more_no_shows += binom.pmf(x, n, p)

print(p_50_or_more_no_shows)
```
